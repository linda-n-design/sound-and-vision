```python
# Fix Duration: Convert "05:03:00" to "5:03"
def fix_duration(duration_str):
    if pd.isna(duration_str) or duration_str == '':
        return None
    
    # Split by colon
    parts = duration_str.split(':')
    
    if len(parts) == 3:
        # First part is minutes, second part is seconds
        minutes = int(parts[0])
        seconds = int(parts[1])
        return f"{minutes}:{seconds:02d}"
    
    return duration_str

# Apply the fix
df['Duration'] = df['Duration'].apply(fix_duration)

# Check the results
print("Fixed duration values:")
print(df['Duration'].head(10))
```

```python
# Fix Duration: Convert datetime.time object to "MM:SS" string
def fix_duration(duration):
    if pd.isna(duration):
        return None
    
    # duration is a datetime.time object
    # The "hour" field actually contains minutes (Apple Music quirk)
    # The "minute" field actually contains seconds
    minutes = duration.hour
    seconds = duration.minute
    return f"{minutes}:{seconds:02d}"

# Apply the fix
df['Duration'] = df['Duration'].apply(fix_duration)

# Check the results
print("Fixed duration values:")
print(df['Duration'].head(10))
print("\nCheck the :59 songs:")
print(df.loc[[105, 107, 192, 215, 234], ['Song', 'Duration']])
```

```python
import json
import pandas as pd
import os

# Check current directory
print("Current directory:", os.getcwd())

# Go up one level to project root (since you're in notebooks/)
os.chdir('..')
print("Changed to:", os.getcwd())

# Now load the JSON
with open('data/songs.json', 'r', encoding='utf-8') as f:
    songs = json.load(f)

df = pd.DataFrame(songs)

# Check what needs splitting
print("\n🔍 Fields with semicolons that need to become arrays:\n")

print("Artist Location:")
print(df[df['Artist Location'].str.contains(';', na=False)][['Song', 'Featured Artist', 'Artist Location']].head())

print("\nGenre:")
print(df[df['Genre'].str.contains(';', na=False)][['Song', 'Genre']].head())

print("\nPrimary Language:")
print(df[df['Primary Language'].str.contains(';', na=False)][['Song', 'Primary Language']].head())

print("\nInstrumental songs (— in language):")
print(df[df['Primary Language'] == '—'][['Song', 'Featured Artist', 'Primary Language']].head(10))
```

```python
# Process each song
for song in songs:
    # Split Artist Location by semicolon
    if song.get('Artist Location') and song['Artist Location'] != '—' and song['Artist Location'] is not None:
        if ';' in song['Artist Location']:
            locations = [loc.strip() for loc in song['Artist Location'].split(';')]
            song['Artist Location'] = locations
    
    # Split Genre by semicolon
    if song.get('Genre') and song['Genre'] != '—' and song['Genre'] is not None:
        if ';' in song['Genre']:
            genres = [g.strip() for g in song['Genre'].split(';')]
            song['Genre'] = genres
    
    # Split Primary Language AND mark instrumentals
    if song.get('Primary Language'):
        if song['Primary Language'] == '—':
            song['Primary Language'] = 'Instrumental'
        elif ';' in song['Primary Language']:
            languages = [lang.strip() for lang in song['Primary Language'].split(';')]
            song['Primary Language'] = languages

# Save updated JSON
with open('data/songs.json', 'w', encoding='utf-8') as f:
    json.dump(songs, f, indent=4, ensure_ascii=False)

print("✅ JSON updated!")

# Verify
df_new = pd.DataFrame(songs)
print(f"\n📊 Summary:")
print(f"  Songs with multiple locations: {df_new['Artist Location'].apply(lambda x: isinstance(x, list)).sum()}")
print(f"  Songs with multiple genres: {df_new['Genre'].apply(lambda x: isinstance(x, list)).sum()}")
print(f"  Songs with multiple languages: {df_new['Primary Language'].apply(lambda x: isinstance(x, list)).sum()}")
print(f"  Instrumental songs: {(df_new['Primary Language'] == 'Instrumental').sum()}")
```

```python
# In Jupyter (or better yet, add this to scripts/process_music.py)
import json

with open('data/songs.json', 'r', encoding='utf-8') as f:
    songs = json.load(f)

# Text fields that should always be strings
text_fields = [
    'Song', 
    'Featured Artist', 
    'Collaborators', 
    'Album', 
    'Cover',
    'Year'  # Even Year should be a string for consistency
]

for song in songs:
    for field in text_fields:
        if song.get(field) and isinstance(song[field], (int, float)):
            song[field] = str(song[field])

# Save
with open('data/songs.json', 'w', encoding='utf-8') as f:
    json.dump(songs, f, indent=4, ensure_ascii=False)

print("✅ Normalized all text fields to strings")
```
