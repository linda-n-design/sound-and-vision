```python
import pandas as pd
```

```python
df = pd.read_excel('../data/music.xlsx')
df.head()
```

```python
# Fix Apple Music duration format
df['Duration'] = df['Duration'].apply(
    lambda x: f"{x.hour}:{x.minute:02d}" if pd.notna(x) else None
)
```

```python
# Export to JSON
df.to_json('../data/songs.json', orient='records', indent=2)
print("JSON exported successfully!")
```

```python
# Check for Postal Service
postal = df[df['Artist'].str.contains('Postal Service', case=False, na=False)]
print(f"Postal Service songs found: {len(postal)}")
print(postal[['Artist', 'Song']].to_string(index=False))

print("\n" + "="*50 + "\n")

# Check for Lush
lush = df[df['Artist'].str.contains('Lush', case=False, na=False)]
print(f"Lush songs found: {len(lush)}")
if len(lush) > 0:
    print(lush[['Artist', 'Song']].to_string(index=False))
else:
    print("No Lush songs found - successfully deleted!")
```
